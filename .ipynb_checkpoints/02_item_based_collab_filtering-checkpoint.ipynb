```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity

# Load datasets
path = 'data/ml-latest-small/'
movies_path = path + "movies.csv"
ratings_path = path + "ratings.csv"

movies = pd.read_csv(movies_path)
ratings = pd.read_csv(ratings_path)

# Step 1: Create an Item-User Matrix
print("\nStep 1: Creating Item-User Matrix...")
item_user_matrix = ratings.pivot(index="movieId", columns="userId", values="rating")
print(item_user_matrix.head())

# Plot the sparsity of the item-user matrix
plt.figure(figsize=(10, 6))
plt.spy(item_user_matrix, markersize=1)
plt.title("Item-User Matrix Sparsity")
plt.xlabel("Users")
plt.ylabel("Movies")
plt.show()

# Step 2: Fill Missing Values with 0
print("\nStep 2: Filling Missing Values with 0...")
item_user_matrix = item_user_matrix.fillna(0)
print(item_user_matrix.head())

# Step 3: Compute Similarity Between Movies
print("\nStep 3: Computing Similarity Between Movies...")
item_similarity = cosine_similarity(item_user_matrix)
similarity_df = pd.DataFrame(item_similarity, index=item_user_matrix.index, columns=item_user_matrix.index)
print("Item Similarity Matrix (first 5 movies):")
print(similarity_df.head())

# Visualize movie similarity heatmap
plt.figure(figsize=(10, 8))
plt.imshow(item_similarity, cmap="viridis", aspect="auto")
plt.colorbar(label="Similarity Score")
plt.title("Movie Similarity Heatmap")
plt.xlabel("Movies")
plt.ylabel("Movies")
plt.show()

# Step 4: Get Similarity Scores for a Target Movie
target_movie_id = 1  # Toy Story (1995) in this dataset
print(f"\nStep 4: Getting Similarity Scores for Target Movie {target_movie_id}...")
target_movie_similarity = similarity_df[target_movie_id].sort_values(ascending=False)
print(target_movie_similarity)

# Plot the similarity distribution
plt.figure(figsize=(10, 6))
target_movie_similarity.plot(kind="bar", color="skyblue")
plt.title(f"Similarity Scores for Movie {target_movie_id}")
plt.ylabel("Similarity")
plt.xlabel("Movies")
plt.show()

# Step 5: Recommend Similar Movies
print("\nStep 5: Recommending Movies Similar to the Target Movie...")
# Exclude the target movie itself
similar_movies = target_movie_similarity.index[1:11]
print(f"Top 10 Similar Movies to Movie {target_movie_id}: {list(similar_movies)}")

# Map movie IDs to titles
movie_titles = movies.set_index("movieId")["title"]
similar_movies_titles = movie_titles.loc[similar_movies]

# Display Recommendations
print(f"\nRecommendations for Movies Similar to {movie_titles[target_movie_id]}:")
print(similar_movies_titles)
```
